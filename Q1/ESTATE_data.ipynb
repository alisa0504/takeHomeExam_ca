```python
digit = {'一': 1, '二': 2, '三': 3, '四': 4, '五': 5, '六': 6, '七': 7, '八': 8, '九': 9}
```

```python
def trans(s):
    num = 0
    if s:
        idx_q, idx_b, idx_s = s.find('千'), s.find('百'), s.find('十')
        if idx_q != -1:
            num  = digit[s[idx_q - 1:idx_q]] * 1000
#             print(num)
        if idx_b != -1:
            num  += digit[s[idx_b - 1:idx_b]] * 100
#             print(num)
        if idx_s != -1:
            num  += digit.get(s[idx_s - 1:idx_s], 1) * 10
#             print(num)
        if s[-1] in digit:
            num  += digit[s[-1]]
#             print(num)
    return num
```

```python
print(trans('三千九百二十一'))
```

```python
df_all[('幾樓','floor_no')] = df_all[('總樓層數', 'total floor number')].str.replace('層','').map(lambda x:trans(x) if type(x)==str else 0 )
```
